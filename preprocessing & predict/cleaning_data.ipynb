```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/Immoweb_Scraper.csv')
df
```

```python
df.head()
```

```python
df = df.drop(["Unnamed: 0"], axis=1)
```

```python
df.drop('isFurnished', axis=1, inplace=True)
```

```python
df=df.replace('Unknown', np.NaN, regex=True)
df.head()
```

```python
df.columns
```

```python
df.dropna(subset=['bedroomCount'], inplace=True)
df.dropna(subset=['netHabitableSurface'], inplace=True)
df.dropna(subset=['terraceSurface'], inplace=True)
df.dropna(subset=['hasTerrace'], inplace=True)
df.dropna(subset=['hasGarden'], inplace=True)
df.dropna(subset=['gardenSurface'], inplace=True)
df.dropna(subset=['hasSwimmingPool'], inplace=True)
df.dropna(subset=['price'], inplace=True)
df.dropna(subset=['kitchen'], inplace=True)
df.dropna(subset=['condition'], inplace=True)
df.dropna(subset=['facadeCount'], inplace=True)
df.isna()
```

```python
df.head()
```

```python
df.count()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.rename(
        columns=({'subtype': 'property-type', 'bedroomCount': 'rooms-number',
                 'netHabitableSurface': 'area', 'fireplaceExists': 'open-fire',
                 'terraceSurface': 'terrace-area', 'hasTerrace': 'terrace',
                 'hasGarden': 'garden ', 'gardenSurface': 'garden-area',
                 'hasSwimmingPool': 'swimming-pool', 'postalCode': 'zip-code', 'locality': 'region',
                 'kitchen': 'equipped-kitchen', 'condition': 'building-state', 'facadeCount': 'facades-number'}),
        inplace=True
)
df
```

```python
df['area']= df['area'].astype(int)
df['rooms-number']= df['rooms-number'].astype(int)
df['zip-code']= df['zip-code'].astype(int)
df['garden-area']= df['garden-area'].astype(int)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.dtypes
```

```python
df = df.drop(df[df["property-type"] == "MANSION"].index)
df = df.drop(df[df["price"] > 9000000].index)
df = df.drop(df[df["rooms-number"] > 50].index)
df = df.drop(df[df["rooms-number"] < 1].index)
df = df.drop(df[df["area"] > 5000].index)
df = df.drop(df[df["terrace-area"] > 200].index)
df = df.drop(df[df["garden-area"] > 600000].index)
```

```python
df.count()
```

```python
df.to_csv('../data/CleanData.csv')
```
